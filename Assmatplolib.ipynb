Scatterplot is a type of graph which is used to represent two sets of data using two different variable. the values are represented through dots which are scattered on the graph

```python
# Q2. ans
plt.scatter(x,y)
plt.xlabel("This is my x Axis")
plt.ylabel("This is my y Axis")
```

Q3 ans.) Subplot is used to plot additional plots , additional axis , additonal rows , in same graph 

```python
# Q3 ans
#line 1
plt.figure(figsize=(10,5))
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x1,y1)
plt.subplot(2,2,1)

#line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x2,y2)
plt.subplot(2,2,2)

#line3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x3,y3)
plt.subplot(2,2,3)

#line4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x4,y4)
plt.subplot(2,2,4)
```

Q3. ans

Bar plot is one of a type of plot used to represent data in bar form the height of the bar represent the value of the data

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

Q4. ans

Box plot is a plot used to represent data form a gap of 5 values, it's yellow shows the median value and the upper 2 whiskeys shows the upper limit and lower wiskeys shows lower limit and the dots represent the values which is far away form teh set of values 

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
np.sort(box1)
plt.boxplot(box1)
plt.show()
```

```python
np.sort(box2)
plt.boxplot(box2)
plt.show()
```
